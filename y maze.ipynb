```python
import pandas as pd
data = pd.read_excel('y maze.xlsx')
```

```python
for i,row in data.iterrows():
    data.loc[i,'total time (s)'] = row['Time in Zone A (s)'] + row['Time in Zone B (s)'] + row['Time in Zone C (s)'] + row['Time in Zone D (s)']
```

```python
data_habituation = data[data['Session'] == 'habituation']
data_habituation = data_habituation.dropna(axis=1, how='all')
data_habituation = data_habituation[['Subject','Treatment','Total Distance (cm)','total time (s)']]
```

```python
data_habituation.head()
```

```python
df_n = data_habituation[data_habituation['Treatment'] == 'Normal'].describe()
df_S = data_habituation[data_habituation['Treatment'] == 'Stress'].describe()
result = pd.concat([df_n, df_S], keys=['Normal', 'Stress'])
result
```

```python
data_test = data[data['Session'] == 'test']
data_test = data_test[['Subject','Treatment','time in tube (s)',
                       'distance_novel_cm','distance_familiar_cm','distance_novel_percentage','distance_familiar_percentage',
                       'Total Distance (cm)','total time (s)',
                 'time_novel_s','time_familiar_s','time_novel_percentage','time_familiar_percentage',
                  'novel first same']]
data_test.reset_index(drop=True, inplace=True)
```

```python
data_test.head()
```

```python
df_t_n = data_test[data_test['Treatment'] == 'Normal'].describe()
df_t_S = data_test[data_test['Treatment'] == 'Stress'].describe()
result_test = pd.concat([df_t_n, df_t_S], keys=['Normal', 'Stress'])
result_test
```

```python
result_test.to_csv('result.csv')
```

```python
#對total distance進行ANOVA，主效果不顯著
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('Q("Total Distance (cm)") ~ C(Treatment) * C(Session)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```

```python
#habituation時兩組的total distance無顯著差異
from scipy import stats
total_distance_normal = data_habituation[data_habituation['Treatment'] == 'Normal']['Total Distance (cm)']
total_distance_stress = data_habituation[data_habituation['Treatment'] == 'Stress']['Total Distance (cm)']

t_statistic, p_value = stats.ttest_ind(total_distance_normal, total_distance_stress)

print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

```python
#treat前後的total time (s)無顯著差異
from scipy import stats

total_time_habituation = data_habituation['total time (s)']
total_time_test = data_test['total time (s)']
t_statistic, p_value = stats.ttest_rel(total_time_test, total_time_habituation)

print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

```python
#normal組treat前後的Total Distance (cm)無顯著差異
from scipy import stats

total_distance_habituation_n = data_habituation[data_habituation['Treatment'] == 'Normal']['Total Distance (cm)']
total_distance_test_n = data_test[data_test['Treatment'] == 'Normal']['Total Distance (cm)']
t_statistic, p_value = stats.ttest_rel(total_distance_test_n, total_distance_habituation_n)

print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

```python
#stress組treat前後的Total Distance (cm)無顯著差異
total_distance_habituation = data_habituation[data_habituation['Treatment']=='Stress']['Total Distance (cm)']
total_dtsance_test = data_test[data_habituation['Treatment']=='Stress']['Total Distance (cm)']
t_statistic, p_value = stats.ttest_rel(total_dtsance_test, total_distance_habituation)

print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Treatment': ['Habituation'] * len(total_distance_habituation) + ['Test'] * len(total_dtsance_test),
    'Total Distance (cm)': pd.concat([total_distance_habituation, total_dtsance_test], ignore_index=True)
})
palette = {"Habituation": "blue", "Test": "green"}
sns.boxplot(x='Treatment', y='Total Distance (cm)', data=df, boxprops={'edgecolor': 'gray'}, palette='RdBu_r')
sns.stripplot(x='Treatment', y='Total Distance (cm)', data=df, color='0.3',alpha=0.5)
plt.savefig('total_distance.png')
```

```python
#兩組在time_novel_percentage上無顯著差異
time_nov_nc = data_test[data_test['Treatment'] == 'Normal']['time_novel_percentage']
time_nov_stress = data_test[data_test['Treatment'] == 'Stress']['time_novel_percentage']

t_statistic, p_value = stats.ttest_ind(time_nov_nc, time_nov_stress)

print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

```python
#兩組在time_novel_s上無顯著差異
time_nov_nc = data_test[data_test['Treatment'] == 'Normal']['time_novel_s']
time_nov_stress = data_test[data_test['Treatment'] == 'Stress']['time_novel_s']

t_statistic, p_value = stats.ttest_ind(time_nov_nc, time_nov_stress)

print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

```python
#兩組在distance_novel_percentage上無顯著差異
dist_nov_nc = data_test[data_test['Treatment'] == 'Normal']['distance_novel_percentage']
dist_nov_stress = data_test[data_test['Treatment'] == 'Stress']['distance_novel_percentage']

t_statistic, p_value = stats.ttest_ind(dist_nov_nc, dist_nov_stress)

print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

```python
# Create a DataFrame for the means
df_means = pd.DataFrame({
    'Group': ['Normal', 'Stress'],
    'Distance in Novel Arm (%)': [dist_nov_nc.mean(), dist_nov_stress.mean()],
    'SEM': [dist_nov_nc.sem(), dist_nov_stress.sem()]
})

# Create a DataFrame for the individual data points
df_data_points = pd.concat([
    pd.DataFrame({'Group': 'Normal', 'Value': dist_nov_nc}),
    pd.DataFrame({'Group': 'Stress', 'Value': dist_nov_stress})
])

# Create bar plot with error bars
sns.barplot(x='Group', y='Distance in Novel Arm (%)', yerr=df_means['SEM'], data=df_means, capsize=.1, palette='coolwarm')

# Add individual data points
sns.stripplot(x='Group', y='Value', data=df_data_points, color='black', size=4, alpha=0.5)

# Add title
plt.title('Control vs. Stress Group - Distance in Novel Arm (%)')

plt.savefig('Control vs. Stress Group - Distance in Novel Arm (%).png')
```

```python
#兩組在distance_novel_cm上有顯著差異
dist_nov_nc = data_test[data_test['Treatment'] == 'Normal']['distance_novel_cm']
dist_nov_stress = data_test[data_test['Treatment'] == 'Stress']['distance_novel_cm']

t_statistic, p_value = stats.ttest_ind(dist_nov_nc, dist_nov_stress)

print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

```python
# Create a DataFrame for the means
df_means = pd.DataFrame({
    'Group': ['Normal', 'Stress'],
    'Distance in Novel Arm (cm)': [dist_nov_nc.mean(), dist_nov_stress.mean()],
    'SEM': [dist_nov_nc.sem(), dist_nov_stress.sem()]
})

# Create a DataFrame for the individual data points
df_data_points = pd.concat([
    pd.DataFrame({'Group': 'Normal', 'Value': dist_nov_nc}),
    pd.DataFrame({'Group': 'Stress', 'Value': dist_nov_stress})
])

# Create bar plot with error bars
sns.barplot(x='Group', y='Distance in Novel Arm (cm)', yerr=df_means['SEM'], data=df_means, capsize=.1, palette='coolwarm')

# Add individual data points
sns.stripplot(x='Group', y='Value', data=df_data_points, color='black', size=4)

# Add significance line
y_max = df_data_points['Value'].max()
plt.plot([0, 0, 1, 1], [y_max + 20, y_max + 30, y_max + 30, y_max + 20], lw=1.5, c='black')
plt.text(0.5, y_max + 30, "*", ha='center', va='bottom', color='black')

# Add title
plt.title('Control vs. Stress Group - Distance in Novel Arm (cm)')

# Show the plot
plt.savefig('Control vs. Stress Group - Distance in Novel Arm (cm).png')
```
